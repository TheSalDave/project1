```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('ProjectDataOriginal.xlsx')
```

```python
df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('HH Electric', fontsize = 40)
red_square = dict(markerfacecolor='r', marker='s')
data = df['HH Electrical (kWh)']
filtered_data = data[~np.isnan(data)]
ax1.boxplot(filtered_data, flierprops=red_square)
ax1.tick_params(axis='both', which='major', labelsize=20)

plt.grid(which = 'major')
plt.xlabel('Timestamp', fontsize = 30)
plt.ylabel('Electrical in kWh', fontsize = 30)
ax2.scatter(x = df['From Timestamp'], y = df['HH Electrical (kWh)'], s = 30)
ax2.tick_params(axis='both', which='major', labelsize=20)

plt.rcParams['figure.figsize'] = [40, 10]

plt.show()
```
