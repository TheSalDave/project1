# Created 3rd column with "transformed" data where I filled in the AVG of the dataset into empty cells as TEMPORARY solution

```python
days = np.zeros((48,365)) #initializing matrix for data storage by days
Data2 = hursleydata.values

for x in range(17520):
    time = x % 48
    day = x // 48
    #filling in datapoints from the dataframe to our matrix by days and time
    days[time][day] = Data2[x]

daySum = np.zeros((1,365))

def sumColumn(m, column):
    total = 0
    for row in range(len(m)):
        total += m[row][column]
    return total

for column in range (365):
    daySum[0][column] = sumColumn(days, column)

```

```python
i = [daySum[0][0], daySum[0][1]]

for x in range (2, 365):
    i.append(daySum[0][x])

plt.plot(i)
plt.title('Days over the year')
```
